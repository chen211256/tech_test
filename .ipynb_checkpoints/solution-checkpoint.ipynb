```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
import numpy as np
from itertools import repeat
from sklearn.metrics import confusion_matrix,mean_absolute_error,mean_squared_error
# LSTM for international airline passengers problem with memory
import tensorflow as tf
from keras.models import Sequential

from keras.layers import Dense,Dropout
from keras.layers import LSTM,Bidirectional

from sklearn.preprocessing import OneHotEncoder,normalize,LabelEncoder
```

```python
m1 = open('data/match_1.json')
df1 = json.load(m1)

m2 = open('data/match_2.json')
df2 = json.load(m2)

df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
```

```python
df1[['label']]=df1[['label']].astype('category')
df2[['label']]=df2[['label']].astype('category')
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
idx1=list(df1['label'].value_counts().index)
idx1
```

```python
def flatten_column(df):
    label_=[]
    norm_=[]
    for i in range(len(df)):
        norm=df.iloc[i,1]
        norm_.append(norm)
        df.iloc[i,0]*len(norm)
        label_.append(list(repeat(df.iloc[i,0],len(norm))))
    
    return flatten(label_),flatten(norm_)
```

```python
lab1_,norm1_=flatten_column(df1)
lab2_,norm2_=flatten_column(df2)
df1_long=pd.DataFrame({'label':lab1_,'norm':norm1_})
df2_long=pd.DataFrame({'label':lab2_,'norm':norm2_})
df1_long
```

```python
# fix random seed for reproducibility
tf.random.set_seed(1234)
```

```python
# Creating a instance of label Encoder.
le = LabelEncoder()


label1 = le.fit_transform(df1_long['label'])


# log normalize
lognorm1_=np.log(norm1_)
df1_long['lognorm']=lognorm1_
```

```python
df1_long['action']=label1
```

```python
print(le.classes_)
le.fit_transform(le.classes_)
```

```python
def reframed_data(dataset,serie_label, start_index, end_index, history_size, target_size):
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i - history_size, i)
        # Reshape data from (history_size,) to (history_size, 1)
        reframed=np.append(dataset[indices],serie_label[indices[-1-target_size]])
        dataset[indices]
        data.append(np.reshape(reframed, (history_size+1, 1)))
        labels.append(dataset[i + target_size])
    print('data size: ',np.array(data).shape, np.array(labels).shape)
    return np.array(data), np.array(labels)
```

```python
plt.figure(figsize=(20,5))
df1_long['lognorm'].plot(subplots=True)
plt.xlim(0,len(df1_long))
```

```python
# The model will be given the last 15 recorded to learn to predict the next time step.
history_size = 10
target_size= 0
split_size = int(0.7* len(df1_long)) 
X_train, y_train= reframed_data(df1_long['lognorm'].values, 
                                          df1_long['action'].values,
                start_index=0, 
                end_index=split_size, 
                history_size=history_size, 
                target_size=target_size)

X_val, y_val= reframed_data(df1_long['lognorm'].values, 
                                      df1_long['action'].values,
                start_index=split_size, 
                end_index=int(0.85* len(df1_long)) , 
                history_size=history_size, 
                target_size=target_size)

X_test, y_test= reframed_data(df1_long['lognorm'].values, 
                                        df1_long['action'].values,
                start_index=int(0.85* len(df1_long)), 
                end_index=None , 
                history_size=history_size, 
                target_size=target_size)
```

```python
batch_size = 5000
buffer_size = 100000
# Batch the data, put them into buffers, shuffle and cache them
train = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train = train.take(batch_size).shuffle(buffer_size).batch(batch_size).cache().repeat()

val= tf.data.Dataset.from_tensor_slices((X_val, y_val))
val = val.take(batch_size).batch(batch_size).cache().repeat()

test = tf.data.Dataset.from_tensor_slices((X_test, y_test))
test = test.take(batch_size).batch(batch_size).cache().repeat()

#.cache() # caches the dataset in memory (avoids having to reapply preprocessing transformations to the input)
#.shuffle(BUFFER_SIZE) # shuffle the samples to have always a random order of samples fed to the network
#.batch(BATCH_SIZE) # batch samples in chunks of size BATCH_SIZE (except the last one, that may be smaller)
#.repeat()) # repeat forever, meaning the dataset will keep producing batches and never terminate running out of data.
```

```python

lstm_model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(8, input_shape=X_train.shape[-2:],return_sequences = True),
    tf.keras.layers.LSTM(4, input_shape=X_train.shape[-2:]),
    tf.keras.layers.Dense(1)])

# Compile the Model
lstm_model.compile(optimizer='adam', loss='mae')


for x, y in val.take(1):
    print(lstm_model.predict(x).shape)

# Train the Model
epochs = 2
steps_per_epoch = 200

trained_lstm = lstm_model.fit(
    train,
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    batch_size=72,
    validation_data=val,
    validation_steps=50)

```
