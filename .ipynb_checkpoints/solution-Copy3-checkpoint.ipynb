```python
def flatten(l):
    return [item for sublist in l for item in sublist]
```

```python
idx1=list(df1['label'].value_counts().index)
idx1
```

```python
def flatten_column(df):
    label_=[]
    norm_=[]
    for i in range(len(df)):
        norm=df.iloc[i,1]
        norm_.append(norm)
        df.iloc[i,0]*len(norm)
        label_.append(list(repeat(df.iloc[i,0],len(norm))))
    
    return flatten(label_),flatten(norm_)
```

```python
lab1_,norm1_=flatten_column(df1)
lab2_,norm2_=flatten_column(df2)
df1_long=pd.DataFrame({'label':lab1_,'norm':norm1_})
df2_long=pd.DataFrame({'label':lab2_,'norm':norm2_})
```

```python
def predict_RF(reg_r,reg_c,X,n):
    X_=[]
    norm=list(X[0:4])
    norm_=[norm]
    action=list(X[5:8])
    action_=[action]
    for i in range(n):
        y_predict_r=reg_r.predict(X.reshape(1,-1))
        y_predict_c=reg_c.predict(X.reshape(1,-1))
        
        if y_predict_c[0]==X[-1]:
            action.append(y_predict_c[0])
            norm.append(y_predict_r[0])
        else:
            action_.append(action)
            norm_.append(norm)
        
        
        X=np.array(flatten([X[1:4],y_predict_r,X[5:8],y_predict_c]))
        X_.append(X)
        
    return X_,action,norm
        
```

```python
def evaluate_preds(y_true, y_pred):
    y_true = tf.cast(y_true, dtype=tf.float32)
    y_pred = tf.cast(y_pred, dtype=tf.float32)
    # Calculate various metrics
    mae = tf.keras.metrics.mean_absolute_error(y_true, y_pred)
    mse = tf.keras.metrics.mean_squared_error(y_true, y_pred) # puts and emphasis on outliers (all errors get squared)
    rmse = tf.sqrt(mse)
    mape = tf.keras.metrics.mean_absolute_percentage_error(y_true, y_pred)

    return {"mae": mae.numpy(),
          "mse": mse.numpy(),
          "rmse": rmse.numpy(),
          "mape": mape.numpy()}
     
```

```python
naive_results = evaluate_preds(y_test[1:],
                               y_pred=naive_forecast)
naive_results
```
